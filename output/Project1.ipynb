        <b>Question 2: The Emotion-Sentiment of Philosophy : Can the texts written by authors and within the schools be classified into varying emotions or general sentiments? If yes, how would that look like?</b></div></html>

<html>
    <div class="alert alert-block alert-info">
<p>The approach towards this problem is by using the pre-existing Python libraries to analyze the sentiments and emotions as presented by the texts. We perform the analysis in two parts. First part performs the Emotion Analysis of 5 authors namely -Epictetus, Descartes , Marcus Aurelius , Wollstonecraft , Berkeley. The emotions are divided into 4 categories: Happy, Sad, Angry, Surprise. Based on the classification of their texts into these emotions, we can try and realize certain characteristics of the authors.</p></div></html>

```python
import text2emotion as te
def detect_emotion(sentence):
    ar=te.get_emotion(sentence)
    return ar
authors=['Epictetus','Descartes','Marcus Aurelius','Wollstonecraft','Berkeley']
for auth in authors:
    df1 = df[df['author'] == auth] 
    numHappy = 0
    numSad = 0
    numAngry = 0
    numSurprise = 0
    for x in range(200):
        emotion = detect_emotion(df1['sentence_lowered'].iloc[x])
        if(emotion['Happy']>0.4):
            numHappy = numHappy + 1
        elif (emotion['Sad']>0.4):
            numSad = numSad + 1
        elif(emotion['Angry']>0.4):
            numAngry = numAngry + 1 
        elif (emotion['Surprise']>0.4):
            numSurprise = numSurprise + 1
    plt.figure(figsize = (7, 7))
    plt.pie([numHappy, numSad, numAngry, numSurprise], labels = ['Happy', 'Sad', 'Angry', 'Surprise'], autopct='%1.2f%%')
    s='Emotional Analysis of Text by: '+auth
    plt.title(s)
    s2 = "Emotion Analysis PieChart - " + auth +".jpg"
    str1 = "../figs/"+ s2
    plt.savefig(str1)
    print(s2, " added to the /figs folder!")
    plt.close()
    #plt.show()
print("\nRun the next cell to display the Pie Charts!")
```

<html>
    <div class="alert alert-block alert-info">
<p>Second part which answers Question 2 performs the Sentiment Analysis of all the 13 school of thoughts. It is divided into 3 categories - Positive, Negative and Neutral. Based on the classification of their texts into these sentiments, we can try and realize the general characteristics of the schools given in the dataset. </p></div></html>

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
def Sentiment_Analysis(sentence):
    sentAnalyzer = SentimentIntensityAnalyzer() 
    sentiment = sentAnalyzer.polarity_scores(sentence)
    if sentiment['compound'] >= 0.05:
        return "positive"
    elif sentiment['compound'] <= -0.05 :
        return "negative"
    else:
        return "neutral"
    
for sch in df['school'].unique():
    df1 = df[df['school'] == sch] 
    post = 0
    negat = 0
    neut = 0
    for i in range (2000):
        sentiment = (Sentiment_Analysis(df1['sentence_lowered'].iloc[i]))
        if sentiment == "positive":
            post += 1
        elif sentiment == "negative":
            negat += 1
        else:
            neut += 1
    plt.figure(figsize = (7, 7))
    plt.pie([post, negat, neut], labels = ['positives', 'negatives', 'neutrals'], autopct='%1.2f%%')
    s='Sentiment Analysis of School- '+sch
    plt.title(s)
    s1="../figs/"+s+".jpg"
    plt.savefig(s1)
    print(s, ".jpg added to the /figs folder!")
    plt.close()
    #plt.show()
print("\nRun the next cell to display the Pie Charts!")
```

        <b>Question 3: The Philosophy of Words : Can the basis of the author's philosophies be realized from their WordCloud representation?</b></div></html>

<html>
    <div class="alert alert-block alert-info">
        <p>A <b>WordCloud</b> helps in highlighting the words that are often used in a given text. Here we analyze the wordcloud of 5 authors - Marx, Plato, Aristotle, Leibniz and Locke. To answer the question, we will look at the larger words (most often used) and compare it with what we know about the authors beliefs and see if they match.</p></div></html>

```python
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
authors=["Marx", "Plato", "Aristotle", "Leibniz", "Locke"]
image =["marx.png", "Plato.png", "aristotle.png","Leibniz.png","locke.png"]
c=0
for auth in authors:
    df1 = df[df.author==auth]
    text = " ".join(txt for txt in df1.sentence_lowered)
    imag = "../data/"+image[c]
    custom_mask = np.array(Image.open(imag))
    c=c+1
    wordcloud = WordCloud(max_font_size=100, max_words=1000,
                          width = 600, height = 400,
                          background_color="white", mask=custom_mask).generate(text)
    image_colors = ImageColorGenerator(custom_mask)     
    wordcloud.recolor(color_func=image_colors)
    plt.figure(figsize=(12,8))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    str1="../figs/wordcloud-"+image[c-1]
    str2= "WordCloud of "+auth
    plt.title(str2)
    plt.savefig(str1)
    #plt.close()
    plt.show()
```
